```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create simple data
X = np.random.rand(200, 2)
y = ((X[:, 0] > 0.5) & (X[:, 1] > 0.5)).astype(int)

# 2. Train on 'safe' data (label 0)
X_safe = X[y == 0]
detectors = []
while len(detectors) < 30:
    d = np.random.rand(2)
    if all(np.linalg.norm(d - s) > 0.1 for s in X_safe):
        detectors.append(d)

# 3. Predict
def predict(x):
    for d in detectors:
        if np.linalg.norm(x - d) < 0.1:
            return 1  # Damaged
    return 0  # Safe

y_pred = np.array([predict(x) for x in X])

# 4. Accuracy
print("Accuracy:", np.mean(y_pred == y))

# 5. Plot
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolor='k')
plt.title("Red = Damaged, Blue = Safe")
plt.show()

********************************************************************

import numpy as np  # For numerical operations and random number generation
import matplotlib.pyplot as plt  # For plotting the result

# 1. Create simple synthetic 2D data
X = np.random.rand(200, 2)  # Generate 200 random data points with 2 features each
y = ((X[:, 0] > 0.5) & (X[:, 1] > 0.5)).astype(int)  # Label = 1 (damaged) if both features > 0.5, else 0 (safe)

# 2. Train the model using only 'safe' data (label 0)
X_safe = X[y == 0]  # Filter out only the safe data points
detectors = []  # Initialize an empty list to store detector points

# Create 30 detectors that are far (more than 0.1 distance) from all safe data points
while len(detectors) < 30:
    d = np.random.rand(2)  # Generate a random 2D detector
    if all(np.linalg.norm(d - s) > 0.1 for s in X_safe):  # Ensure it's not close to any safe point
        detectors.append(d)  # Add to detector list

# 3. Define prediction function using trained detectors
def predict(x):
    for d in detectors:
        if np.linalg.norm(x - d) < 0.1:  # If point is close to any detector, label it as damaged
            return 1
    return 0  # Else, label it as safe

# Predict labels for all data points using the trained model
y_pred = np.array([predict(x) for x in X])

# 4. Calculate and display accuracy
print("Accuracy:", np.mean(y_pred == y))  # Compare predicted vs actual labels

# 5. Plot the classified data points
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolor='k')  # Color points based on prediction
plt.title("Red = Damaged, Blue = Safe")  # Add plot title
plt.show()  # Display the plot

***************************************************************************************

import numpy as np
import matplotlib.pyplot as plt

THEORY:
We import the required libraries. numpy helps in creating arrays and doing math operations, while matplotlib.pyplot is used to visualize results with graphs.

1. Create simple data
X = np.random.rand(200, 2)
y = ((X[:, 0] > 0.5) & (X[:, 1] > 0.5)).astype(int)

THEORY:
We generate 200 random 2D points between 0 and 1.
We assign a label: if both values in a point are greater than 0.5, it's considered damaged (label = 1), else it's safe (label = 0).

2. Train on 'safe' data (label 0)
X_safe = X[y == 0]

THEORY:
We filter the data to keep only "safe" points (where label is 0). These are used to generate detectors.

detectors = []
while len(detectors) < 30:
d = np.random.rand(2)
if all(np.linalg.norm(d - s) > 0.1 for s in X_safe):
detectors.append(d)

THEORY:
We create 30 detectors.
Each detector is a random 2D point that must be at least 0.1 distance away from all safe data points.
This ensures that detectors only cover non-self (potential damage) space.

3. Predict
def predict(x):
for d in detectors:
if np.linalg.norm(x - d) < 0.1:
return 1 # Damaged
return 0 # Safe

THEORY:
This function checks if a new point is close to any detector.
If it is within 0.1 distance from a detector, it is classified as damaged (1).
Otherwise, it's safe (0).

y_pred = np.array([predict(x) for x in X])

THEORY:
We apply the prediction function to all points in our dataset and store the results in y_pred.

4. Accuracy
print("Accuracy:", np.mean(y_pred == y))

THEORY:
We compare predicted labels (y_pred) with actual labels (y).
We calculate the average number of correct predictions to get the accuracy.

5. Plot
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolor='k')
plt.title("Red = Damaged, Blue = Safe")
plt.show()

THEORY:
We plot all data points in 2D space.
Points are colored based on predicted labels: red for damaged (1), blue for safe (0).
This helps visually confirm how well our model is classifying.

************************************************************

Here is the **simplified version of the code with an easy-to-understand explanation of the theory** tailored for explaining to an external examiner:

---

###  **THEORY (Simplified)**

**Artificial Immune System (AIS)** is a machine learning method inspired by how our bodys immune system works:

* It **learns to detect abnormalities** just like white blood cells recognize viruses.
* Used here for **structural damage classification**, e.g., predicting if a structure is safe or damaged using sensor data.

**How AIS works (in simple terms):**

1. **Self vs. Non-self**: It learns what normal looks like (safe structure).
2. **Detectors**: It creates random points (detectors) that stay away from normal data (non-self = possible damage).
3. **Prediction**: If new data is close to a detector, its abnormal (damaged); else, its safe.
4. This model is called the **Negative Selection Algorithm (NSA)**  a type of AIS.

---

###  **EXPLANATION FOR EXTERNAL VIVA**

You can say:

> In this AIS-based approach, I generated synthetic data representing structural conditions.
I applied the **Negative Selection Algorithm**, where only 'safe' data is used to train the model.
It then creates 'detectors' which are points that stay away from this safe data. During prediction, 
if a new point falls near any detector, it is classified as 'damaged'. 
This mimics how our immune system detects non-self (foreign) elements. 
The model achieved decent accuracy and also visualized the classification zones clearly.

******************************************************************


###  **Point-Wise Explanation of Theory (Easy Technical Language)**

1. **What is AIS (Artificial Immune System)?**
   AIS is a computer algorithm inspired by the **human immune system**.
   Just like our body detects viruses (non-self) and ignores normal cells (self),
   AIS is used to **detect abnormal patterns** in data  like **structural damage**.

2. **What is NSA (Negative Selection Algorithm)?**
   It is a type of AIS algorithm that **learns only from normal (safe) data**
   and then creates "detectors" that identify anything that doesnt look normal.

3. **How is it useful for Structural Damage Classification?**
   We simulate a situation where the model:

   * learns what safe structures look like (self),
   * then flags unknown patterns (non-self) as potential damage.

4. **What kind of data is used?**
   We create **synthetic 2D data** where each point has 2 features (like vibration values),
   and we label it as:

   * **0 = safe** (both values  0.5),
   * **1 = damaged** (both values > 0.5).

5. **How are detectors used?**
   Detectors are random points that:

   * stay away from all safe data,
   * if a new point is **close to any detector**, it is classified as **damaged**.

6. **How do we test the model?**

   * We split data into train and test sets.
   * Train the model only on safe data.
   * Test it using new points.
   * We calculate **accuracy** and show results with a **color plot** (red = damaged, blue = safe).

7. **Why AIS is useful?**

   * Can work in **noisy** or uncertain environments.
   * **Adapts** like the immune system.
   * Doesnt need full knowledge of all damage types  can detect unknown anomalies.

---

###  **What to Tell Examiner in Practical**

> "I implemented the Artificial Immune System using the Negative Selection Algorithm.
> I generated 2D synthetic data where points with both values > 0.5 are considered damaged.
> The model learns from safe data and creates detectors that dont match safe data.
> Then during testing, if a new point lies close to a detector, it is classified as damaged.
> I visualized the results and achieved decent accuracy using this simple immune-based method."

---

###  **Expected Questions from External Examiner**

| Question                                                    | Suggested Answer                                                                                                                                           |
| ----------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **1. What is the inspiration behind AIS?**                  | AIS is inspired by the biological immune system, where the body distinguishes between self and non-self to fight infections.                               |
| **2. What is NSA?**                                         | NSA (Negative Selection Algorithm) is an AIS technique that detects anomalies by training only on normal data and flagging anything that deviates from it. |
| **3. Why did you use synthetic data?**                      | To simulate real-world structural conditions simply and clearly for learning and demonstration purposes.                                                   |
| **4. Why only safe data is used for training?**             | Because in NSA, detectors are trained to avoid self (safe) data  so they can recognize non-self (damaged) data later.                                     |
| **5. What happens if a point is close to a detector?**      | It is classified as damaged (non-self).                                                                                                                    |
| **6. What are the advantages of AIS?**                      | It handles noise well, adapts over time, and doesnt need to see all types of damage beforehand.                                                           |
| **7. Any real-life use of this?**                           | Yes, AIS is used in fault detection, cybersecurity, and structural health monitoring in bridges, aircraft, etc.                                            |
| **8. What did you visualize in the output?**                | I showed predicted results: red points as damaged, blue as safe, and detector locations in green.                                                          |
| **9. Why is this method better than basic classification?** | Because it can detect unknown damage types and is more robust to unseen anomalies.                                                                         |

---

************************************************************************


### **AIM**

To implement an Artificial Immune System (AIS) using randomly generated synthetic data.

---

### **PROBLEM STATEMENT**

To apply artificial immune pattern recognition techniques to perform a task of **structural damage classification** using synthetic data inputs.

---

### **THEORY**

**Artificial Immune Systems (AIS)** are biologically inspired computational models that replicate the working principles of the human immune system. AIS is widely used in areas such as **pattern recognition**, **anomaly detection**, **classification**, and **optimization**.

In this practical, AIS is applied to the task of **structural damage classification**, which is a critical part of **Structural Health Monitoring (SHM)**. SHM ensures the safety and operational integrity of structures such as bridges, buildings, and aircraft by detecting and classifying damage conditions accurately and early.

#### Key Concepts in AIS:

1. **Clonal Selection**:
   High-performing solutions (antibodies) are cloned and mutated to explore better possibilities.

2. **Negative Selection**:
   Detectors are generated to avoid known (self) patterns and detect unknown (non-self) or abnormal patterns  especially useful in anomaly detection.

3. **Immune Memory**:
   Previously encountered patterns are remembered for faster and more accurate future recognition.

#### AIS Workflow for Damage Classification:

* Structural data is preprocessed (can be from sensors or simulated).
* Feature vectors are formed to represent structural conditions.
* AIS is applied to learn and evolve a classification model.
* New structural inputs are classified into categories such as:

  * **Healthy**
  * **Minor Damage**
  * **Severe Damage**

The AIS algorithm adapts over time, mimicking the dynamic and learning behavior of natural immune systems, making it robust for practical, real-time applications.

---

### **ADVANTAGES**

* **Robust to Noise**:
  Negative selection makes the system effective in noisy or uncertain data conditions.

* **Adaptive Learning**:
  AIS learns and remembers new patterns over time, enhancing prediction accuracy with experience.

* **Parallelism**:
  Inspired by immune responses, AIS can be parallelized for faster processing and scalability.

* **Nonlinear Problem Solving**:
  Capable of handling complex relationships in structural data that traditional linear models may miss.

---

### **DISADVANTAGES**

* **Computational Complexity**:
  Some AIS algorithms may become computationally expensive, especially with large or high-dimensional data.

* **Parameter Sensitivity**:
  AIS performance can depend heavily on parameters like clone size, mutation rate, or radius of detection  requiring tuning.

* **Scalability Limitations**:
  Without optimization, AIS may struggle with very large datasets or real-time applications due to resource constraints.

```
